## Profit VS Discount

```python
from matplotlib import style
style.use('dark_background')
plt.figure(figsize=(10,6))
plt.scatter(df['Sales'],df['Profit'])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

## Profit VS Quantity

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

## profit/loss and sales in segment

```python
df.groupby("Segment")[['Profit','Sales']].sum().plot.bar(color=["red","blue"],figsize=(8,5),ylabel="profit/loss vs sales")
```

## segment wise sales in each region

```python
plt.figure(figsize=(8,10))
plt.title("segment wise sales in each region")
sns.barplot(x="Region",y="Sales",hue="Segment",data=df,order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel("Region",fontsize=10)
plt.show()
```

## profit/loss and sales in region

```python
df.groupby('Region')[["Profit","Sales"]].sum().plot.bar(color=["red","blue"],figsize=(8,5))
plt.ylabel("profit/loss and sales")
plt.show()
```

## profit/loss and sales across states

```python
df.groupby('State')[["Profit","Sales"]].sum().sort_values(by="Sales",ascending=False).plot.bar(color=["red","blue"],figsize=(8,5))
plt.ylabel("profit/loss and sales")
plt.xlabel("states")
plt.ylabel("profit/loss and sales")
plt.title

plt.show()
```

## profit/loss and sales by category

```python
df.groupby('Category')[["Profit","Sales"]].sum().plot.bar(color=["red","blue"],figsize=(8,5))
plt.ylabel("profit/loss and sales")
plt.show()
```
